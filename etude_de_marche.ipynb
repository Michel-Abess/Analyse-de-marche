### Implémentation du cluster hierarchique

```python
# creation du model de clustering hierarchique
hiercluster = AgglomerativeClustering(affinity='euclidean', linkage="ward", compute_full_tree=True)
```

```python
# Adaptation du model aux données
hiercluster.set_params(n_clusters=5)
clusters = hiercluster.fit_predict(df_country_scaled)
np.bincount(clusters) # nombre d'individus de chaque clusters
```

```python
print(len(clusters))
```

```python
# transformation de la matrice en dataframe
df_country_scaled_clustered = pd.DataFrame(df_country_scaled, columns=df_country_data.columns, index=df_country_data.index)
df_country_scaled_clustered.head()
```

```python
# Associons chaque individu au cluster auquel il appartient - 2
df_country_scaled_clustered['cluster'] = clusters
df_country_scaled_clustered.head()
```

```python
df_country_scaled_clustered['cluster'].unique()
```

```python
# Affichage du dendogram
from scipy.cluster.hierarchy import linkage, fcluster
Z = linkage(df_country_scaled_clustered, 'ward') # 
names = main_df['Zone'].values

plt.figure(figsize=(10,25))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('distance')
dendrogram(Z, labels=names, color_threshold=14, orientation='left')
plt.axvline(x=14, c='grey', lw=1, linestyle='dashed')
plt.show
plt.savefig('dendogram')
```

```python
import numpy as np
from scipy import stats
```

```python
stats.ks_2samp(main_df['Evolution population (%) 2019/2018'],
list(np.random.normal(np.mean(main_df['Evolution population (%) 2019/2018']),
                        np.std(main_df['Evolution population (%) 2019/2018']), 1000)))
```

Le risque de rejetter à tord l'hypothèse H0(superposition potentielle de la loi normale theorique correspondante à la distribution empirique est elevé).
On ne peut donc pas rejetter l'hypothèse de normalité au niveau de test de 5% pour la variable 'Evolution population (%)2019/2018'

conclusion: la variable 'Evolution population (%) 2019/2018' est gaussienne. 

- Test sur de normalité sur la variable : Valeur PIB par hbts (US $)

```python
stats.ks_2samp(main_df['Valeur PIB par hbts (US $)'],
list(np.random.normal(np.mean(main_df['Valeur PIB par hbts (US $)']),
                        np.std(main_df['Valeur PIB par hbts (US $)']), 1000)))
```

la P-value < 5% pour la variable 'Valeur PIB par hbts (US $)'. 
les chances de réalisation de l'hypothèse H0 sont faibles
(rappel : moyenne et variannce équivalence pour la distribution
empirique et la distribution normale théorique correspondante)

conclusion: le PIB par hbts n'obéit pas à une loi normale.

- Test de l'égalité des variances

```python
scipy.stats.bartlett(main_df_cluster_4, main_df_cluster_2)
```

L\'erreur lié au rejet à tord de H0 (hypothèse d'égalité des variances) est faible. On peut considérer que les variances sont différentes. On peut conclure à ce niveau que les deux distributions sont donc différentes.

verifions quand même l'égalité des moyennes pour aller plus loin.

```python
print(scipy.stats.ttest_ind(main_df_cluster_2, main_df_cluster_4,
                      equal_var=False))
print("\n")
print(f"l\'erreur lié au rejet à tord de H0 est faible\
 les moyennes ne sont pas egales pour les deux\
 distributions")
```
